# Notebook 1 - Heterogenous Causal Effects of In-Hospital Insulin on 30-Day Readmission

## 1. Introduction

### 1.1 Background and Policy Context

Hospital readmission within 30 days of discharge is one of the most closely 
watched quality metrics in the US healthcare system. Since 2012, the Hospital 
Readmissions Reduction Program (HRRP) has penalised hospitals financially for 
excessive 30-day readmission rates under Medicare, which is the American federal insurance 
programme covering patients over 65 and with disabilities, as a way to incentivize the recourse to effective treatment procedures. 
For hospitals in the US, reducing readmissions is  therefore not only a clinical goal but a direct economic imperative.

Diabetic patients account for a disproportionate share of hospital readmissions. 
Managing diabetes during a hospital stay is complex: blood sugar levels fluctuate 
with illness, stress, and changes in diet, and poorly controlled glycaemia during 
the stay can delay recovery and increase the risk of complications after discharge. 
Insulin is the most common diabetes treatment prescribed during hospital stays, 
used in over half of all diabetic inpatient encounters in our dataset. Whether 
insulin prescription relative to no insulin or to the use of different medications actually reduces the probability 
of readmission within the 30-days window specifically, after properly accounting for differences in 
patient severity, is a question with direct implications for prescribing protocols 
and hospital performance on a financially consequential metric.

### 1.2 The Identification Challenge

Answering this question from observational data is not straightforward. Physicians 
do not randomly assign insulin. They tend to prescribe it to their sickest patients, those 
with the worst glycaemic control, the highest medication burden, and the most 
complex clinical picture. Those same patients are inherently more likely to be 
readmitted, regardless of what they receive. In the raw data, insulin patients 
therefore appear to have *worse* readmission outcomes than non-insulin patients. This may 
not be because insulin is harmful or ineffective, but because it is given to the patients who were 
always at highest risk.

This is the **confounding-by-indication** problem we are trying to address in this project. It means that a naive 
comparison of readmission rates between insulin and non-insulin patients does not 
measure the effect of the drug, it only measures the effect of being sicker. Standard 
OLS regression, even with controls, is vulnerable to this problem when the 
confounders are high-dimensional and their relationship with treatment and outcome 
is nonlinear.

This project addresses this problem using **Double Machine Learning (DML)**, which 
combines flexible machine learning nuisance models with Neyman-orthogonal scores 
and cross-fitting to remove confounding from a rich set of clinical covariates. We first show the naive OLS 
estimate, then add controls, then move to DML, to show exactly how the estimated effect changes as identification improves. The gap 
between the naive and DML estimates quantifies how much 
confounding-by-indication distorts naive analysis in this setting.

### 1.3 Research Question

**What is the causal effect of in-hospital insulin prescription on 30-day readmission for diabetic patients, and does this effect vary across patient subgroups defined by age, gender, and prior hospitalisation history?**


We study two estimands:

- **ATE** = E[Y(1) − Y(0)]: the average effect of insulin across all patients 
  in the analytic sample
- **ATT** = E[Y(1) − Y(0) | D=1]: the average effect among patients who 
  actually received insulin

The ATE and ATT answer the average-effect question. We then try to answers 
the heterogeneity question: does insulin help all patients equally, or does the 
benefit concentrate in specific subgroups? This matters practically because if insulin 
is only effective for certain patient profiles, that has direct implications for 
how prescribing decisions should be targeted from the point of view of both patients and hospitals.

### 1.4 PICO

We formalise the research question using the PICO framework, standard in 
clinical research for specifying the population, intervention, comparison, 
and outcome of interest.

| Element | Definition |
|---|---|
| **P - Population** | Diabetic inpatients at 130 US hospitals, 1999–2008. One observation per patient. Hospice and deceased discharges are excluded as they are not at risk of readmission. |
| **I - Intervention** | Any in-hospital insulin prescription during the stay. |
| **C - Comparison** | No in-hospital insulin prescription. |
| **O - Outcome** | Binary indicator equal to 1 if the patient was readmitted to hospital within 30 days of discharge, 0 otherwise. |


### 1.5 Data

The analysis uses the **UCI Diabetes 130-US Hospitals dataset**, originally 
collected and described by Strack et al. (2014) in their study of the impact 
of HbA1c measurement on hospital readmission rates. The dataset covers 
**130 US hospitals** over a **ten-year period from 1999 to 2008**, and contains 
detailed information on patient demographics, diagnosis codes, laboratory 
results, medications prescribed during the stay, and readmission outcomes.

Strack et al. (2014) studied a different research 
question, namely the effect of HbA1c measurement on readmission, using primarily 
descriptive and logistic regression methods. Our project uses the same dataset 
but asks a different causal question and applies a different methodology.

The raw dataset contains **101,766 hospital encounters**.
